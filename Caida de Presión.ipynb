```python
import numpy as np
import matplotlib.pyplot as plt

# Matplotlib configuration (optional)
plt.rcParams.update({
    "text.usetex": False,
    "font.family": "serif",
    "mathtext.fontset": "cm",
})

# Given parameters
Qcc = 21.60         # Flow rate in cc/s
k = 5300            # Constant (Pa.s^n)
n = 0.378           # Flow behavior index
Lgallo = 60.00      # Total pipe length in mm

Dgalloent = 4.00    # Inlet diameter in mm
Dgallosal = 6.75    # Outlet diameter in mm

# Define the inlet and outlet radii (in mm)
r_in_mm = Dgalloent / 2.0   # mm
r_out_mm = Dgallosal / 2.0   # mm

# Total pipe length in SI unit (meters)
L_total_m = Lgallo / 1000.0    # m

# Convert the flow rate from cc/s to m^3/s (1 cc = 1e-6 m^3)
Q_m3 = Qcc * 1e-6              # m^3/s

# List of N values to iterate over
N_values = [2]

# Function to compute the local radius in mm at position x (in m)
def local_radius_mm(x):
    return r_in_mm + (r_out_mm - r_in_mm) * (x / L_total_m)

# Initialize the plot
plt.figure(figsize=(10, 5))

# Variables to track the maximum pressure and corresponding time
max_pressure = 0
max_time = 0

# Loop over each N value
for N in N_values:
    # Create an array for the spatial coordinate x (in m)
    x_array = np.linspace(0, L_total_m, N)
    dx = x_array[1] - x_array[0]

    # Prepare arrays for acumulative time and acumulative pressure drop
    time_array = np.zeros_like(x_array)       # Time in seconds
    pressure_array = np.zeros_like(x_array)     # Pressure in Pa

    # Integration along the pipe (using the midpoint for better accuracy)
    for i in range(1, N):
        # Define previous and current position
        x_prev = x_array[i-1]
        x_curr = x_array[i]
        
        # Midpoint for better approximation
        x_mid = 0.5 * (x_prev + x_curr)
        
        # Local radius in mm and convert it to meters
        R_mid_mm = local_radius_mm(x_mid)   # mm
        R_mid_m = R_mid_mm / 1000.0           # m
        
        # Compute local cross-sectional area at the midpoint (in m^2)
        A_mid = np.pi * (R_mid_m)**2
        
        # Compute local velocity: v = Q / A (m/s)
        v_mid = Q_m3 / A_mid
        
        # Time increment for the small spatial segment
        dt = dx / v_mid
        
        # Convert dx (m) to mm for the pressure drop formula (1 m = 1000 mm)
        dx_mm = dx * 1000.0
        
        # Calculate the term inside the bracket in the pressure drop equation.
        # Note: (R_mid_mm*1e-3) converts the radius to meters for the cubic term.
        term = ((3 + 1/n) * Qcc * 1e-6) / (np.pi * (R_mid_mm * 1e-3)**3)
        
        # Differential pressure drop (dP) for this small segment, in Pa.
        # Using the formula: dP = 2k * (dx_mm / R_mid_mm) * (term ** n)
        dP = 2 * k * (dx_mm / R_mid_mm) * (term ** n)
        
        # Accumulate the time and pressure drop
        time_array[i] = time_array[i-1] + dt
        pressure_array[i] = pressure_array[i-1] + dP

    # Convert cumulative pressure drop to MPa for plotting (1 MPa = 1e6 Pa)
    pressure_array_MPa = pressure_array * 1e-6

    # Plot Cumulative Pressure versus Time for this N
    plt.plot(time_array, pressure_array_MPa, lw=2, label=f'N = {N-1}')

    # Check if this is the maximum pressure
    if pressure_array_MPa[-1] > max_pressure:
        max_pressure = pressure_array_MPa[-1]
        max_time = time_array[-1]

# Add a point for the maximum pressure value
plt.scatter(max_time, max_pressure, color='red', zorder=5)
plt.text(max_time, max_pressure, f'Max: {max_pressure:.2f} MPa', fontsize=10, ha='right')

# Finalize the plot
plt.xlabel('Tiempo (s)', fontsize=14)
plt.ylabel('Caida de Presión (MPa)', fontsize=14)
plt.title('Cantidad de Divisiones', fontsize=16)
#plt.xlim(0, 2.4)
plt.ylim(0, 4.75)
#plt.xticks(np.arange(0, 2.5, 0.5))
plt.yticks(np.arange(0, 4.75, 1))
plt.grid(True)
plt.legend(loc='upper left')
plt.savefig('dP Gallo n=1.png', dpi=100, bbox_inches='tight', transparent=False)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.mplot3d.art3d import Poly3DCollection

def create_cylinder(start, end, radius=0.5, resolution=20):
    # Vector in direction of axis
    v = end - start
    
    # Make unit vector
    mag = np.linalg.norm(v)
    v = v / mag
    
    # Make basis vectors
    not_v = np.array([1, 0, 0])
    if (v == not_v).all():
        not_v = np.array([0, 1, 0])
    
    n1 = np.cross(v, not_v)
    n1 /= np.linalg.norm(n1)
    n2 = np.cross(v, n1)
    
    # Surface of cylinder
    t = np.linspace(0, mag, 2)
    theta = np.linspace(0, 2*np.pi, resolution)
    t, theta = np.meshgrid(t, theta)
    
    x = start[0] + (n1[0] * np.cos(theta) + n2[0] * np.sin(theta)) * radius + v[0] * t
    y = start[1] + (n1[1] * np.cos(theta) + n2[1] * np.sin(theta)) * radius + v[1] * t
    z = start[2] + (n1[2] * np.cos(theta) + n2[2] * np.sin(theta)) * radius + v[2] * t
    
    return x, y, z

def create_rectangular_cylinder(start, end, width, height):
    # Vector in direction of axis
    v = end - start
    
    # Make unit vector
    mag = np.linalg.norm(v)
    v = v / mag
    
    # Make basis vectors
    not_v = np.array([1, 0, 0])
    if (v == not_v).all():
        not_v = np.array([0, 1, 0])
    
    n1 = np.cross(v, not_v)
    n1 /= np.linalg.norm(n1)
    n2 = np.cross(v, n1)
    
    # Define the vertices of the rectangular prism
    vertices = []
    # Bottom face
    vertices.append(start - (width/2)*n1 - (height/2)*n2)  # 0
    vertices.append(start + (width/2)*n1 - (height/2)*n2)  # 1
    vertices.append(start + (width/2)*n1 + (height/2)*n2)  # 2
    vertices.append(start - (width/2)*n1 + (height/2)*n2)  # 3
    # Top face
    vertices.append(end - (width/2)*n1 - (height/2)*n2)    # 4
    vertices.append(end + (width/2)*n1 - (height/2)*n2)    # 5
    vertices.append(end + (width/2)*n1 + (height/2)*n2)    # 6
    vertices.append(end - (width/2)*n1 + (height/2)*n2)    # 7
    
    # Define the faces using vertex indices
    faces = [
        [vertices[0], vertices[1], vertices[2], vertices[3]],  # bottom
        [vertices[4], vertices[5], vertices[6], vertices[7]],  # top
        [vertices[0], vertices[1], vertices[5], vertices[4]],  # front
        [vertices[2], vertices[3], vertices[7], vertices[6]],  # back
        [vertices[0], vertices[3], vertices[7], vertices[4]],  # left
        [vertices[1], vertices[2], vertices[6], vertices[5]]   # right
    ]
    
    return faces

# Example points
x = [0, 0, 0, 12, 15, 65]
y = [0, 0, 63, 63, 63, 63]
z = [60, 0, 0, 0, 0, 0]

# Define custom radius for each connecting line
radius = [Rgallo, Rcpri1, Rcsec, Rpuente, Hpart]  # One less than number of points

# Create 3D plot
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Create scatter plot
scatter = ax.scatter(x, y, z, c=z, cmap='turbo')

# Add colorbar
#plt.colorbar(scatter)

# Labeling
ax.set_xlabel('Eje X (mm)')
ax.set_ylabel('Eje Y (mm)')
ax.set_zlabel('Eje Z (mm)')
ax.set_title('3D Plot')

# Create cylinders between points
for i in range(len(x)-1):
    start = np.array([x[i], y[i], z[i]])
    end = np.array([x[i+1], y[i+1], z[i+1]])
    
    # Connecting line
    ax.plot([x[i], x[i+1]], 
            [y[i], y[i+1]], 
            [z[i], z[i+1]], 
            color='red', linestyle='--')
    
    if i == len(x)-2:  # Last segment
        # Rectangular cylinder
        width = 11.5  # Customize these values
        height = 2.0
        faces = create_rectangular_cylinder(start, end, height, width)
        
        # Create poly collection
        rect_prism = Poly3DCollection(faces, alpha=0.3)
        rect_prism.set_facecolor('blue')
        rect_prism.set_edgecolor('black')
        ax.add_collection3d(rect_prism)
    else:
        # Regular cylinder
        cx, cy, cz = create_cylinder(start, end, radius=radius[i])
        ax.plot_surface(cx, cy, cz, alpha=0.3, color='blue')

# Get the limits of the plot
x_lim = ax.get_xlim()
y_lim = ax.get_ylim()
z_lim = ax.get_zlim()

# Calculate the range of each dimension
x_range = x_lim[1] - x_lim[0]
y_range = y_lim[1] - y_lim[0]
z_range = z_lim[1] - z_lim[0]

# Find the maximum range
max_range = max(x_range, y_range, z_range)

# Set new limits to make the scaling equal
ax.set_xlim([np.mean(x_lim) - max_range/2, np.mean(x_lim) + max_range/2])
ax.set_ylim([np.mean(y_lim) - max_range/2, np.mean(y_lim) + max_range/2])
ax.set_zlim([np.mean(z_lim) - max_range/2, np.mean(z_lim) + max_range/2])

# Force equal aspect ratio
ax.set_box_aspect((1,1,1))

# Set the view angle
ax.view_init(elev=20, azim=25)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.mplot3d.art3d import Poly3DCollection

def create_cylinder(start, end, radius=0.5, resolution=20, length_steps=50):
    v = end - start
    mag = np.linalg.norm(v)
    v = v / mag
    not_v = np.array([1, 0, 0])
    if np.allclose(v, not_v):
        not_v = np.array([0, 1, 0])
    n1 = np.cross(v, not_v)
    n1 /= np.linalg.norm(n1)
    n2 = np.cross(v, n1)

    t = np.linspace(0, mag, length_steps)
    theta = np.linspace(0, 2 * np.pi, resolution)
    t, theta = np.meshgrid(t, theta)

    x = start[0] + (n1[0] * np.cos(theta) + n2[0] * np.sin(theta)) * radius + v[0] * t
    y = start[1] + (n1[1] * np.cos(theta) + n2[1] * np.sin(theta)) * radius + v[1] * t
    z = start[2] + (n1[2] * np.cos(theta) + n2[2] * np.sin(theta)) * radius + v[2] * t

    return x, y, z, t

def create_rectangular_prism(start, end, width, height):
    v = end - start
    mag = np.linalg.norm(v)
    v = v / mag
    not_v = np.array([1, 0, 0])
    if np.allclose(v, not_v):
        not_v = np.array([0, 1, 0])
    n1 = np.cross(v, not_v)
    n1 /= np.linalg.norm(n1)
    n2 = np.cross(v, n1)

    vertices = []
    vertices.append(start - (width/2)*n1 - (height/2)*n2)  # 0
    vertices.append(start + (width/2)*n1 - (height/2)*n2)  # 1
    vertices.append(start + (width/2)*n1 + (height/2)*n2)  # 2
    vertices.append(start - (width/2)*n1 + (height/2)*n2)  # 3
    vertices.append(end - (width/2)*n1 - (height/2)*n2)    # 4
    vertices.append(end + (width/2)*n1 - (height/2)*n2)    # 5
    vertices.append(end + (width/2)*n1 + (height/2)*n2)    # 6
    vertices.append(end - (width/2)*n1 + (height/2)*n2)    # 7

    faces = [
        [vertices[0], vertices[1], vertices[2], vertices[3]],  # bottom
        [vertices[4], vertices[5], vertices[6], vertices[7]],  # top
        [vertices[0], vertices[1], vertices[5], vertices[4]],  # front
        [vertices[2], vertices[3], vertices[7], vertices[6]],  # back
        [vertices[0], vertices[3], vertices[7], vertices[4]],  # left
        [vertices[1], vertices[2], vertices[6], vertices[5]]   # right
    ]

    return faces

def get_color(pressure_val, p_min, p_max):
    norm_pressure = (pressure_val - p_min) / (p_max - p_min)
    return plt.cm.rainbow(norm_pressure)

def draw_gradient_rectangular_cylinder(ax, start, end, p1, p2, width, height, p_min, p_max, n_steps=20):
    for i in range(n_steps):
        frac1 = i / n_steps
        frac2 = (i + 1) / n_steps
        seg_start = start + (end - start) * frac1
        seg_end = start + (end - start) * frac2
        seg_pressure = p1 + (p2 - p1) * frac1
        color = get_color(seg_pressure, p_min, p_max)
        color = list(color)
        color[3] *= 0.6  # reduce alpha

        faces = create_rectangular_prism(seg_start, seg_end, width, height)
        rect = Poly3DCollection(faces, facecolor=color, edgecolor='none')
        ax.add_collection3d(rect)

# === Input values ===
x = [0, 0, 0, 12, 15, 65]
y = [0, 0, 63, 63, 63, 63]
z = [60, 0, 0, 0, 0, 0]

pressure = [round((dPgallo + dPcpri + dPcsec + dPpuente + dPpart)*1e-6, 2), round((dPgallo + dPcpri + dPcsec + dPpuente)*1e-6, 2), round((dPgallo + dPcpri + dPcsec)*1e-6, 2), round((dPgallo + dPcpri)*1e-6, 2), round((dPgallo*1e-6), 2), 0.1]
print(pressure)

p_min = min(pressure)
p_max = max(pressure)

radius = [Rgallo, Rcpri1, Rcsec, Rpuente, Hpart]  # per-segment radius

# === Plot setup ===
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(x, y, z, c=pressure, cmap='turbo')
plt.colorbar(scatter, label='Presión de inyección (MPa)')

ax.set_xlabel('Eje X (mm)')
ax.set_ylabel('Eje Y (mm)')
ax.set_zlabel('Eje Z (mm)')
ax.set_title('Presión al final del llenado (MPa)')

# === Create cylinders ===
for i in range(len(x) - 1):
    start = np.array([x[i], y[i], z[i]])
    end = np.array([x[i+1], y[i+1], z[i+1]])
    p1, p2 = pressure[i], pressure[i+1]

    ax.plot([x[i], x[i+1]], [y[i], y[i+1]], [z[i], z[i+1]], color='black', linestyle='--')

    if i == len(x) - 2:
        width = 11.5
        height = 2.0
        draw_gradient_rectangular_cylinder(ax, start, end, p1, p2, height, width, p_min, p_max, n_steps=20)
    else:
        cx, cy, cz, t = create_cylinder(start, end, radius=radius[i], resolution=30, length_steps=50)
        pressures_along = np.linspace(p1, p2, cx.shape[1])
        norm_pressures = (pressures_along - p_min) / (p_max - p_min)
        colors = plt.cm.rainbow(norm_pressures)
        colors[:, 3] *= 0.6

        for j in range(cx.shape[1] - 1):
            color_slice = np.tile(colors[j], (cx.shape[0], 2, 1))
            ax.plot_surface(
                cx[:, j:j+2], cy[:, j:j+2], cz[:, j:j+2],
                facecolors=color_slice,
                shade=False,
                linewidth=0,
                antialiased=False
            )

# === Adjust view ===
x_lim, y_lim, z_lim = ax.get_xlim(), ax.get_ylim(), ax.get_zlim()
x_range, y_range, z_range = x_lim[1] - x_lim[0], y_lim[1] - y_lim[0], z_lim[1] - z_lim[0]
max_range = max(x_range, y_range, z_range)
ax.set_xlim([np.mean(x_lim) - max_range/2, np.mean(x_lim) + max_range/2])
ax.set_ylim([np.mean(y_lim) - max_range/2, np.mean(y_lim) + max_range/2])
ax.set_zlim([np.mean(z_lim) - max_range/2, np.mean(z_lim) + max_range/2])
ax.set_box_aspect([1, 1, 1])
ax.view_init(elev=20, azim=25)

#plt.tight_layout()
plt.show()
```
